Temps d'exécution de l'inférence en ms pour une évidence

```python
import matplotlib.pyplot as plt

# Données
nb_nodes = [1, 3, 10, 20]
time_inference = [
    [10, 20],
    [76800, 79556, 86009, 70713],
    [80881, 70009, 86301, 71781],
    [69980, 71530, 78269, 95093],
    [68135, 91594, 85614, 83655]
]

# Création du graphique
plt.figure(figsize=(4, 4))

for i in range(len(nb_nodes)):
    plt.plot(time_inference[i], label=f'{nb_nodes[i]} nodes')

plt.xlabel('Nombre de parents')
plt.ylabel('Temps d\'exécution (ms)')
plt.title('Temps d\'exécution de l\'inférence en ms pour une évidence')
plt.legend()
plt.show()
```
